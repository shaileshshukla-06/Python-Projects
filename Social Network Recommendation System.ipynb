```python
import json
 
```

```python
# Load the JSON file
def load_data(filename):
    with open(filename, "r") as file:
        data = json.load(file)
    return data
```

```python
data = load_data("data.json")
```

```python
# Write a function to dispaly user and their connections
def display_users(data) :
    print("Users and their connections : \n")
    
    for user in data['users'] :
        print(f"ID:{user['id']}. {user['name']} is friend with : {user['friends']} and like pages {user['liked_pages']}")

    print("\nPages Information :")
    for page in data['pages'] :
        print(f"{page['id']} : {page['name']}")

display_users(data)
```
